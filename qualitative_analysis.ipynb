```python
def show_file(filename):
    with open(filename, 'rb') as file:
        grads = pickle.load(file)
    print(grads.shape)
    # x = np.abs(grads) ** 0.85
    # x = grads ** 2
    # x /= x.sum(axis=1, keepdims=True)

    frame = pd.DataFrame(grads, columns='pos vel theta omega'.split())
    sns.kdeplot(data=frame, bw_adjust=3)
    return frame.head()
```

```python
show_file("default.pkl")
```
